```python
import json
data = pd.read_csv('data.csv')
json_data = data.to_json(orient='records')
pretty_json = json.dumps(json.loads(json_data), indent=4, ensure_ascii=False)
#write to pretty_json json file
with open('data1.json', 'w') as f:
    f.write(pretty_json)
```
